```python
from datasets import load_dataset
raw_datasets = load_dataset("code_search_net", "python")
```

```python
raw_datasets["train"]
```

```python
print(raw_datasets["train"][123456]["whole_func_string"])
```

```python
training_corpus = (
    raw_datasets["train"][i : i + 1000]["whole_func_string"]
    for i in range(0, len(raw_datasets["train"]), 1000)
)
```

```python
def get_training_corpus():
    return (
        raw_datasets["train"][i : i + 50000]["whole_func_string"]
        for i in range(0, len(raw_datasets["train"]), 50000)
    )


training_corpus = get_training_corpus()
```

```python
from transformers import AutoTokenizer
old_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
old_tokenizer.tokenize(raw_datasets["train"][123456]["whole_func_string"])
```

```python
# training a new tokenizer from the old one with 50k samples

tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, 50000)
```

```python
tokenizer.save_pretrained("code-search-net-tokenizer")
```

```python
tokenizer.tokenize(raw_datasets["train"][123456]["whole_func_string"])
```
